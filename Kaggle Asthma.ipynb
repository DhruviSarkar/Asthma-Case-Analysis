```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("AsthmaDataset.csv")

print("Column Names:")
print(df.columns)
```

```python
# Sample dataset
data = {
    "Age Group": ["0–4", "5–14", "15–19", "20–24", "25–34", "35–64", "65+"],
    "Number of Cases": [64500, 203200	, 164300, 208400, 234700, 384400, 197600]  # Replace with actual data
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(df["Age Group"], df["Number of Cases"], color="skyblue", edgecolor="black")
plt.title("Asthma Cases by Age Group", fontsize=16)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Number of Cases", fontsize=14)
plt.yticks(range(0, max(df["Number of Cases"]) + 50000, 50000))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

# Display the graph
plt.show()

```

```python
data = {
    "Region": ["Northeast", "Midwest", "South", "West"],
    "Children": [90900, 104200, 160900, 119900],  
    "Adults": [287500, 260700, 377400, 304400]    
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set positions for the bars
x = range(len(df["Region"]))
width = 0.4  # Width of the bars

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar([p - width/2 for p in x], df["Children"], width=width, label="Children", color="skyblue", edgecolor="black")
plt.bar([p + width/2 for p in x], df["Adults"], width=width, label="Adults", color="salmon", edgecolor="black")

# Add labels, title, and legend
plt.title("Asthma Cases by Region and Age Group", fontsize=16)
plt.xlabel("Region", fontsize=14)
plt.ylabel("Number of Cases", fontsize=14)
plt.xticks(x, df["Region"], fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Display the graph
plt.tight_layout()
plt.show()
```

```python
data = {
    "Ratio of Family Income to Poverty Threshold": ["0-0.99", "1.00-2.49", "2.50-4.49", "4.50 and above"],
    "Number of Cases": [249100, 341600, 295500, 364200],  
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.bar(df["Ratio of Family Income to Poverty Threshold"], df["Number of Cases"], color="purple", edgecolor="black")
plt.title("Asthma Cases by Ratio of Family Income to Poverty Threshold", fontsize=16)
plt.xlabel("Ratio of Family Income to Poverty Threshold", fontsize=14)
plt.ylabel("Number of Cases", fontsize=14)
plt.yticks(range(0, max(df["Number of Cases"]) + 50000, 50000))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

```
